## **fix the column names**

```python
#fix cloumn names
df.columns = df.columns.str.strip()
print(df.columns)
```

## **droping the columns**

```python
df.drop(columns = ['GitHub Profile Link'], inplace= True)
```

```python
df.columns
```

### **type of identification**

```python
df.dtypes
```

### **fix age**

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df
```

### **time stamp**

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp']).dt.tz_localize(None)
print(df['Timestamp'])
```

### **cgpa columns**

```python
cgpa_cols = [
    'SSC CGPA',
    'Intermediate / Diploma CGPA',
    'Current CGPA'
]
for col in cgpa_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
print(df[cgpa_cols])
```

### **covert into cgpa upto 10**

```python
for col in cgpa_cols:
    df.loc[df[col] > 10, col] = df[col] / 100
print(df[cgpa_cols])
```

### **cgpa states**

```python
df['Current CGPA'].agg(['min', 'max' , 'mean'])
```

### **placement package**


```python
df[['LPA_Min', 'LPA_Max']] = (
    df['Expected Placement Package (LPA)']
    .str.split('-', expand=True)
)
df['LPA_Min'] = pd.to_numeric(df['LPA_Min'], errors='coerce')
df['LPA_Max'] = pd.to_numeric(df['LPA_Max'], errors='coerce')
df['Expected_LPA_Avg'] = (df['LPA_Min'] + df['LPA_Max']) / 2
df
```

```python
df.columns
```

### **fixed study hours**

```python
study_map = {
    'Less than 1 hour': 0.5,
    '1-2 hours': 1.5,
    '2-4 hours': 3,
    'More than 4 hours': 5
}
df['Average Daily Study Hours'] = df['Average Daily Study Hours'].map(study_map)
df['Average Daily Screen Time'] = df['Average Daily Screen Time'].map(study_map)
df
```
