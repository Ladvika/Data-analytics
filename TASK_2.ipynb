## Clean and preprocess a dataset

```python
# Handling Missing Values
df["Price"].fillna(df["Price"].median(), inplace=True)

# Remove Duplicates
df.drop_duplicates(inplace=True)

# Formatting (Date)
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df
```

# Perform slicing, filtering, sorting, and subsetting operations

```python
# Column Selection
df[["Product", "Quantity", "Price"]]

# Row Filtering
df[df["Quantity"] > 3]

# Multiple Conditions
df[(df["Region"] == "North") & (df["Price"] > 20000)]

# Sorting
df.sort_values(by="Price", ascending=False)


df
```

## Apply aggregating functions using groupby

```python
df.groupby("Product")["Price"].mean()

df.groupby("Region").agg({
    "Quantity": "sum",
    "Price": "mean"
})
df
```

## Merge and join two datasets

```python
# Create Inventory Dataset
inventory = pd.DataFrame({
    "Product": ["Laptop", "Mobile", "Tablet", "Headphones"],
    "Stock": [50, 120, 80, 200]
})

# Merge
merged_df = pd.merge(df, inventory, on="Product", how="left")
df
```

## Create pivot tables and reshape datasets using melt() and pivot()

```python
# Pivot Table
pd.pivot_table(
    df,
    values="Price",
    index="Region",
    columns="Product",
    aggfunc="mean"
)

# Melt (Wide → Long)
pd.melt(
    df,
    id_vars=["Order_ID", "Region"],
    value_vars=["Quantity", "Price"]
)
```

### Perform descriptive statistical analysis (mean, median, std, IQR)

```python
df["Price"].mean()
df["Price"].median()
df["Price"].std()

Q1 = df["Price"].quantile(0.25)
Q3 = df["Price"].quantile(0.75)
IQR = Q3 - Q1
IQR
```

### Create basic plots: line, bar, scatter, histogram.

```python
import matplotlib.pyplot as plt

# Line plot
df.groupby("Order_Date")["Quantity"].sum().plot()
plt.show()

# Bar plot
df["Product"].value_counts().plot(kind="bar")
plt.show()

# Scatter plot
plt.scatter(df["Quantity"], df["Price"])
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.show()

# Histogram
df["Price"].hist(bins=20)
plt.show()
```

## Visualize a dataset using heatmap, boxplot and pairplot

```python
import seaborn as sns

# Boxplot
sns.boxplot(x=df["Price"])
plt.show()

# Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()

# Pairplot
sns.pairplot(df[["Quantity", "Price"]])
plt.show()
```
