STUDENT DATA

**STEP 1.Load the Dataset**

```python
import pandas as pd
import numpy as np
```

```python
# insert the csv student file
df = pd.read_csv('/content/students_data.csv')
```

```python
#few rows seleted
df.head()
```

```python
#understand the struture of the data
df.info()
```

**STEP 2.	Explore the Data**

```python
#shape
df.shape
```

```python
if 'name' in df.columns:
    #split name into frist and last name
    df['frist_name'] = df['name'].astype(str).str.split().str[0]
    df['last_name'] = df['name'].astype(str).str.split().str[1]
    #drop the name colums
    df.drop(columns=['name'], inplace=True)
```

```python
df
```

```python
df = df.rename(columns={'student_id':'ID','frist_name':'Frist_Name','last_name':'Last_Name','age':'Age','gender':'Gender','grade':'Class','math_score':'Math_Score','english_score':'English_Score','science_score':'Science_Score','enrolled_date':'Enrolled_Date','remarks':'Grade'})
df
```

```python
#fill the nan with age values
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age).astype(int)
df
```

```python
# Summary statistics
df.describe(include='all')
```

```python
df['Gender'] = df['Gender'].astype(str).str.strip().str.lower()

df['Gender'] = df['Gender'].replace({
    'female': 'F',
    'f': 'F',
    'male': 'M',
    'm': 'M'
})
```

```python
df
```

```python
df['Class'] = df['Class'].astype(str).str.strip().str.lower()

df['Class'] = df['Class'].replace({
    '10': 'X',
    '10th': 'X',
    '11': 'XI',
    '11th': 'XI',
    '12': 'XII',
    '12th': 'XII'
})
```

```python
df
```

```python
df['Math_Score'] = df['Math_Score'].fillna(df['Math_Score'].mean())
df['Math_Score'] = df['Math_Score'].astype(int)
```

```python
df
```

```python
# Replace 'missing' with NaN
df['English_Score'] = df['English_Score'].replace('missing', np.nan)

# Convert to numeric
df['English_Score'] = pd.to_numeric(df['English_Score'], errors='coerce')

# Fill missing values with mean
df['English_Score'] = df['English_Score'].fillna(df['English_Score'].mean())

# Convert to integer
df['English_Score'] = df['English_Score'].astype(int)
```

```python
df
```

```python
df['Enrolled_Date'] = pd.to_datetime(
    df['Enrolled_Date'],
    errors='coerce',
    dayfirst=True,
    format='mixed'
)

df['Enrolled_Date'] = df['Enrolled_Date'].dt.strftime('%d-%m-%Y')
```

```python
df
```

```python
# Clean Grade column
df['Grade'] = df['Grade'].astype(str).str.strip().str.lower()

# Standardize Grade values
df['Grade'] = df['Grade'].replace({
    'excellent': 'O',
    'good': 'A',
    'good student': 'A',
    'average': 'B',
    'needs improvement':'C',

    'poor': 'D'
})
```

```python
df
```

```python
cols = df.columns.tolist()

# Take last two columns
last_two = cols[-2:]

# Remaining columns except last two
remaining = cols[:-2]

# New column order: 1st column + last two + rest
new_order = [remaining[0]] + last_two + remaining[1:]

# Reorder DataFrame
df = df[new_order]
```
